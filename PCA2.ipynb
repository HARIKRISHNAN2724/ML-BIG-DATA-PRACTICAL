```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
rng = np.random.RandomState(1)
X = np.dot(rng.rand(2,2),rng.randn(2,200)).T
rng
X
```

```python
# ORGINAL SHAPE OF DATA.
X.shape
```

```python
D = pd.DataFrame(X,columns=['Z','Y'])
D
```

```python
#PLOT THE DATA 
sns.set()
sns.lmplot(x='Z',y='Y',data=D)
```

```python
sns.lmplot(x='Z',y='Y',data=D,fit_reg=False)
```

```python
from sklearn.decomposition import PCA
model = PCA(n_components=1)
model
```

```python
model.fit(D)
X_2D = model.transform(D)
X_2D
```
